### Covid data is obtained from [covid19.org](https://api.covid19india.org/)

```python

import requests 
url = "https://api.covid19india.org/v4/data.json"

# a request is being sent to the web page
response = requests.get(url)
```

```python
# response 200 means the page is responsive, like 404 means the page doesnot exist
response
```

```python
# displaying content of the page
response.content
```

```python
import json 
data = json.loads(response.content)
```

```python
data
```

```python
data_lst = []
err_states = []


for state in data.keys():
    try:
        for name, d1 in data[state]['districts'].items():
            res = {'state': state, 'district': name, **d1['total']}
            data_lst.append(res)
    except Exception:
        err_states.append(state)
```

```python
import pandas as pd
df = pd.DataFrame(data_lst)
# df1 = pd.DataFrame(err_states)

df.head()
```

### Want to join the two data frames based on district column
